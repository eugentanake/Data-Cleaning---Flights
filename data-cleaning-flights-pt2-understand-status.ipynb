```python
df[df['status']=='Delayed']['number'].value_counts()
```

```python
tempList = []
for i in list(df[df['status']=='Delayed']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'Delayed' in list(group[1]['status']):
            #display((group[1]))
            tempList.append(list(group[1]['status']))
print(pd.Series(tempList).value_counts())
```

```python
tempList = []
for i in list(df[df['status']=='GateClosed']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'GateClosed' in list(group[1]['status']):
            #display((group[1]))
            tempList.append(list(group[1]['status']))
print(pd.Series(tempList).value_counts())
```

```python
tempList = []
for i in list(df[df['status']=='Boarding']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'Boarding' in list(group[1]['status']):
            #display((group[1]))
            tempList.append(list(group[1]['status']))
print(pd.Series(tempList).value_counts())
```

```python
tempList = []
for i in list(df[df['status']=='Expected']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'Expected' in list(group[1]['status']):
            #display((group[1]))
            tempList.append(list(group[1]['status']))
print(pd.Series(tempList).value_counts())
```

The codes above checks the flight numbers that has the statuses 'Delayed','GateClosed','Boarding','Expected' and group them by planned departure timing. We can see from the results that we have a unique row for "Boarding" and 10 unique rows for "Expected", the rest of the flights has subsequently departed and can be removed.

```python
tempList = []
for i in list(df[df['status']=='Boarding']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'Boarding' in list(group[1]['status']):
            #display((group[1]))
            if len(list(group[1]['status']))==1:
                tempList.append(i)
df[df['number'] == tempList[0]]
```

```python
tempList = []
for i in list(df[df['status']=='Expected']['number'].value_counts().index):
    for group in df[df['number']==i].groupby('departure.scheduledTimeLocal'):
        if 'Expected' in list(group[1]['status']):
            #display((group[1]))
            if len(list(group[1]['status']))==1:
                tempList.append(i)
for i in tempList:
    display(df[df['number'] == i])
```
