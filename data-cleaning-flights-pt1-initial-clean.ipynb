This was part of an assignment for a data visualization course which I thought could be used to display some visualization.

- Target: The target was to show destinations for parents to bring their children overseas as it was the June Holidays.

The data was obtained from aerodatabox api. However as a free member there was a limit to how much data I could obtain. To be able to obtain a week's worth of flights I had to obtain the data twice. So there are duplicates which I have to clean out first.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df1 = pd.read_csv('/kaggle/input/flights-from-aerodataboxapi/flight_list.csv')
df2 = pd.read_csv('/kaggle/input/flights-from-aerodataboxapi/flight_list2.csv')
df1.head()
```

```python
len(df1)
```

```python
df1.isnull().sum()
```

The columns with thousands of missing values would not be of immediate help to the analysis so I will exclude them.

As for the Unnamed: 0 column, it was created during the API extraction process so it is okay to remove it.

```python
df1 = df1.iloc[:, 1:18]
```

```python
df1.head()
```

```python
print(df1["arrival.quality"].unique())
print(df1["departure.quality"].unique())
print(df1["departure.checkInDesk"].unique())
print(df1["isCargo"].unique())
```

The data in Utc timing will not be of help as Local timing is available.

The data in arrival and departure quality will also not be of help as well as it does not make much sense.

As the purpose is to show flights people would take, the cargo flights are excluded from the api download.

However I will keep checkInDesk for now as it may potentially be of help.

```python
df1 = df1.drop(["arrival.quality","departure.quality","departure.actualTimeUtc","departure.scheduledTimeUtc","isCargo"], axis = 1)
```

```python
df1.head()
```

```python
# Do the same with the second half collected
df2.head()
```

```python
print(len(df2))
df2.isnull().sum()
```

```python
df2 = df2.iloc[:, 1:18]
print(df2["arrival.quality"].unique())
print(df2["departure.quality"].unique())
print(df2["departure.checkInDesk"].unique())
print(df2["isCargo"].unique())
```

```python
df2 = df2.drop(["arrival.quality","departure.quality","departure.actualTimeUtc","departure.scheduledTimeUtc","isCargo"], axis = 1)
df2.head()
```

```python
df = pd.concat([df1,df2])
len(df)
```

```python
df = df.drop_duplicates(keep='first')
```
