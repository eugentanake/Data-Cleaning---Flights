The following was done on Jupyter notebook as it contained a personal code with limited usage.

```python
import pandas as pd
import requests
```

```python
dt = pd.date_range(start='2022-10-25 00:00:00', end='2022-10-30', freq='12H')
dt # yyyy-mm-dd
```

If I were to decrease the frequency to 1 day, it would exceed the limit imposed on the free account and only give me about 70% of the data of that day so I increased the frequency. Better to obtain duplicates than to miss out on data.

```python
dtdate = []
dthour = []
dtcombined = []

for i in dt:
    dtdate.append(i.strftime('%Y-%m-%dT'))
    dthour.append(i.strftime('%H'))
for date, hour in zip(dtdate,dthour):
    dtcombined.append(date+hour+":00")
dtcombined
```

```python
endUrls = []
for i in range(len(dtcombined)-1):
    endUrls.append(dtcombined[i]+"/"+dtcombined[i+1])
endUrls
```

```python
baseurl = "https://aerodatabox.p.rapidapi.com/flights/airports/icao/WSSS/"

querystring = {"withLeg":"true","direction":"Departure","withCancelled":"true","withCodeshared":"true","withCargo":"false","withPrivate":"false","withLocation":"false"}

headers = {
    "X-RapidAPI-Key": "", # I have a free code that is limited so i removed it here
    "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}
dataList = []

for i in endUrls:
    url = baseurl+i
    response = requests.request("GET", url, headers=headers, params=querystring)
    data = response.json()
    tempdf = pd.json_normalize(data,'departures')
    dataList.append(tempdf)
```

```python
df = pd.concat(dataList)
```

The final df is saved to csv file for further use.
